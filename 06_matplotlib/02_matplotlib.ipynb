Part One: Understanding the Figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating random data

x = np.arange(0, 10)

y = 2*x
```

```python
# plotting
plt.plot(x, y)
```

```python
# NOTE: As you can see that it returns object details with memory location and the image. So, to ignore the object we can add semicolon (;) or add plt.show() command at the end.

# NOTE: Also keep in mind that if you use .py files instead of .ipynb (jupyter notebook) then add plt.show() command at the end compulsorly.

plt.plot(x, y);
```

```python
# trying other things like labels, x and y limitations and titles

# NOTE: Keep in mind that you make changes in same cell.

plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Y=2*X')

plt.xlim(0, 10)
plt.ylim(0, 20);
```
