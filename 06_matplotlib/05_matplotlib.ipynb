## Part Three: Figure Parameters

```python
# NOTE: To remove some blurriness while creating canvas we can use 'dpi'(dots per inch) parameter while creating figure.

import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(0, 10, 11)
b = a**4

x = np.arange(0, 10)
y = 2*x
```

```python
# fig = plt.figure(dpi=100)

# fig = plt.figure(figsize=(10, 3), dpi=100)
# fig = plt.figure()

fig = plt.figure(figsize=(2, 2), dpi=200)

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b);
```

```python
# NOTE: On the alternate you can do as providing figsize parmeter also to increase the canvas size. As done above.

# to save figures
fig.savefig('new_figure.png')

# NOTE: You can notice that due to our parameter values the size of the image is bigger.
```
