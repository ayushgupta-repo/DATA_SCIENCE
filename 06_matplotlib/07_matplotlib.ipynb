## Matplotlib Styling - Legends

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating datasets to show legends

x = np.linspace(0, 11, 10)
y = 2*x

a = np.arange(0, 10)
b = a**4
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.plot(a, b);
```

```python
# NOTE: As you can see that different bars must be represented with different colors but still there are some problem like which bar shows which type of equation representation.

plt_fig = plt.figure()

plt_axes = plt_fig.add_axes([0, 0, 1, 1])

plt_axes.plot(x, y, label='X vs 2*X')
plt_axes.plot(a, b, label='A vs A**4')

plt_axes.legend();
```

```python
plt_figure_1 = plt.figure()

plt_figure_1_axes = plt_figure_1.add_axes([0, 0, 1, 1])

plt_figure_1_axes.plot(x, y, label='X vs 2**X')
plt_figure_1_axes.plot(a, b, label='A vs A**4')

# We can modify the location of the legends with the loc parameter and we can provide numeric location (consider documentation) and the word notations as well as shown below
# plt_figure_1_axes.legend(loc='upper left');

# plt_figure_1_axes.legend(loc='upper center');

# NOTE: You can also do manual locaiting by passing the fractional values in tuple form as in loc

plt_figure_1_axes.legend(loc=(1.1, 0.5));
```
