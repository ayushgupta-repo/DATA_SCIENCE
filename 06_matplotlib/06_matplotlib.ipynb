## Matplotlib - Subplots Functionality

NOTE: As in previous we are able to create side by side multiple plots using axes. But there is a pre-defined functionality that is subplots() that do all those multiple axes stuff.

Subplots() allows us to easily create figures and axes objects in side by side functionalities.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating datasets
a = np.linspace(0, 10, 11)
b = a**4

x = np.arange(0, 10)
y = 2*x
```

```python
# using subplots method
fig, axes = plt.subplots()
axes.plot(x, y)
```

```python
# creating multiple axes and plotting points to it

fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
type(axes)
```

```python
axes

# NOTE: Now you can see that dimensions get modified. So, you can use indexing to do multiple stuffs, which we will discuss later.
```

```python
axes.shape # Returns shape in tuple form
```

```python
# Using indexing to plot in the axes created as shown below

# creating another figure due to some problem

fig, mul_axes = plt.subplots(nrows=1, ncols=2)

mul_axes[0].plot(x, y)
mul_axes[1].plot(a, b)

# NOTE: You can notice the spacing between the axes is somewhat not good. We will further discuss upon this as well.
```

```python
# NOTE: By changing the sizes of plotting statment creates error due to change in dimensions (in case of it changes from current dimension to some other dimension only).

# creating axes to show above problem

# creating multi-dimensional axes
multi_dim_fig, multi_dim_axes = plt.subplots(nrows=2, ncols=2)

multi_dim_fig[0].plot(x, y)
multi_dim_fig[1].plot(a, b)

# NOTE: It will throw error because now the dimensions got changed due to creation of multi_dimensional_axeses. So, in order to solve this do plotting based on dimensions as we do in normal multi-dimensional arrays.
```

```python
# NOTE: First get to know about the dimension first as shown below.

multi_dim_axes.shape
```

```python
# OR you can do like calling the var_axes

multi_dim_axes # Returns list of axes and on the basis of that you can find the dimension.
```

```python
# As it is currently now 2*2 dimension then we use multi-dimensional indexing for plotting, titling and providing limitations.

multi_dim_axes[0][0].plot(x, y)
multi_dim_axes[1][1].plot(a, b);

fig # You can do if some problem occurs during viewing the plottings.
```

```python
# most popular dimensional of subplotting is (nrows=3, ncols=1)

pop_fig, pop_axes = plt.subplots(nrows=3, ncols=1)

# In case of this (means having only one column) you can do looping as shown below

for ax in pop_axes:
    ax.plot(x, y)
```

```python
# NOTE: As you can see in above that there are some spacing problems. So, we can solve this using one method (tight_layout()) and manual spacing using wspace (width space) and hspace(height space).

tig_lay_fig, tig_lay_axes = plt.subplots(nrows=2, ncols=2)

tig_lay_axes[0][0].plot(x, y)
tig_lay_axes[1][1].plot(a, b)

# Now performing tight_layout 

plt.tight_layout() # Automatically some spacing gets provided
```

```python
# NOTE: Now doing manual spacing using wspace and hspace. Keep in mind that they provide spaces between axeses and they are in fractions.

tig_lay_fig.subplots_adjust(wspace=1, hspace=1)

tig_lay_fig

# NOTE: You can play around with these parameters for more enhanced plotting.
```

```python
# Now we are doing labelling, limits and title using previous axes plotting methods (set_method)

tig_lay_axes[0][0].set_ylabel('Y LABEL 0,0')
tig_lay_axes[0][0].set_xlim(2, 6)

tig_lay_axes[1][1].set_title('TITLE 1,1')

# Similary you can do more with others as well as per the need.
```

```python
tig_lay_fig
```

```python
# NOTE: Giving title to whole axes as a single one also known as figure_level_title and provide various font parameters (like fontstyle, fontsize and many more) as shown below

tig_lay_fig.suptitle('Figure Level', fontsize=16)
```

```python
plt.tight_layout()

tig_lay_fig
```

```python
# saving plot

tig_lay_fig.savefig('new_subplots.png', bbox_inches='tight')
```

```python
# NOTE: Making some changes in figure to get the proper figsize like changing the figure width as shown below

tig_lay_fig.set_figwidth(10)

tig_lay_fig
```

```python
tig_lay_fig.savefig('new_subplots_1.png', bbox_inches='tight')
```

```python
# Try more things like figsize and dpi as done previous by creating new fig and axes

try_fig, try_axes = plt.subplots(nrows=2, ncols=2, figsize=(2, 2), dpi=200)
plt.tight_layout()

try_axes[0][0].plot(x, y)
try_axes[1][0].plot(a, b)
```
