## Matplotlib - Part Two - Implementing Figures and Axes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating data

a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0, 10)

y = 2*x
```

```python
x
```

```python
y
```

```python
# creating figures and then using axes (with modifications using set_modification_name)

fig = plt.figure() # creates the blank canvas of particular default size

# NOTE: We can provide figsize as well to modify the canvas size

axes = fig.add_axes([0, 0, 1, 1]) # creating axes

# NOTE: Syntax be like fig.add_axes([lower_x, lower_y, widht, height])

axes.plot(x, y);
```

```python
# plotting another figure with scratch

fig = plt.figure()

axes2 = fig.add_axes([0, 0, 1, 1])

axes2.plot(a, b);
```

```python
# NOTE: Play around with the add_axes parameter to get better hand on that.

type(fig) # to get fig type
```

```python
# Creating multiple axes (LARGE and SMALL)

another_fig = plt.figure()

# LARGE AXES
axes1 = another_fig.add_axes([0, 0, 1, 1])

# setting labels, title and limits for larger axes
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_title('Power of 4')

axes1.plot(a, b);

# SMALL AXES
axes2 = another_fig.add_axes([0.2, 0.5, 0.25, 0.25])
# axes2 = another_fig.add_axes([1, 1, 2, 2]) # For trying purpose only to show we can not only limit ourselve within 0 to 1 we can go beyond as well (especially in case of legends plotting which we learn later).

# setting labels, title and limits for smaller axes
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_title('Power of 4')
axes2.plot(a, b);
```
