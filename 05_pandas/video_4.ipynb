## DataFrame - Part Three - Working with columns

```python
# You can access columns
df['total_bill']
```

```python
df.head()['total_bill']
```

```python
# NOTE: Keep in mind that each column of dataframe is nothing but a Series sharing common rows
type(df['total_bill'])
```

```python
# You can also grab specific number of columns by creating list and passing it will accessing

mycols = ['total_bill', 'tip']
#df[mycols]
# OR
df[['total_bill', 'tip']]
```

```python
# You can peform operation using different columns
df['tip'] + df['total_bill']
```

```python
df['tip'] / df['total_bill']
```

```python
100 * df['tip']/df['total_bill'] # Mathematical calculation for finding percentage

df
```

```python
# Adding new column

df['tip_percentage'] = 100 * df['tip']/df['total_bill']
```

```python
df.head()
```

```python
# Adding new column that already exists just overrides the value contained by the column.

df['price_per_person'] = df['total_bill'] / df['size']
```

```python
df.head()
```

```python
# Rounding off OR Selecting digits after decimal
# np.round(value, digits_after_decimal)

df['price_per_person'] = np.round(df['price_per_person'], 2)
```

```python
df.head()
```

```python
# Droping columns

# Keep in mind that axis=0 means dropping by finding provided name in row (by default)
# axis = 1 means dropping by finding provided name in column

# So below code will show KeyError because it will search in labeled index rather than searching in row
df.drop('tip_percentage')
```

```python
# So you need to provide axis=1 to search key in column

df.drop('tip_percentage', axis=1)
```

```python
# NOTE: Also but you notice that originally it is not dropped from the df

df.head()
```

```python
# So in order to make changes happen to original df there are two ways:
# Way-1: In drop there is one attribute inplace make it True by default it is False (But due to version changes avoid this way)

# Way-2: Assigning df to returning value of drop like shown below

df = df.drop('tip_percentage', axis=1)
```

```python
df.head()
```

```python
df.shape # As in return value you can see it is (row, column) so the axis=0 means row and axis=1 means column
```
